```python
# Function to recommend users based on common cities and skills
def recommend_users(user_id, users_data):
    # Extract the city and skills of the given user
    user_city = users_data.loc[user_id, 'CITY']
    user_skills = users_data.loc[user_id, 'SKILLS']

    # Filter users with the same city and different user_id
    city_match = users_data[users_data['CITY'] == user_city]
    city_match = city_match[city_match.index != user_id]

    # Filter users with at least one common skill
    skill_match = users_data[users_data['SKILLS'].apply(lambda x: any(skill in user_skills for skill in x.split()))]
    skill_match = skill_match[skill_match.index != user_id]

    # Combine both city and skill matches to get final recommendations
    recommendations = pd.concat([city_match, skill_match]).drop_duplicates()

    return recommendations

# Example: Recommend users for the user with user_id 0
user_id_to_recommend = 0
recommendations = recommend_users(user_id_to_recommend, users_data)

# Display the recommendations
print(f"Recommendations for user {user_id_to_recommend}:\n")
print(recommendations)
```

# Common cities

```python
# Function to recommend users based on common cities
def recommend_users_by_city(user_id, users_data):
    # Extract the city of the given user
    user_city = users_data.loc[user_id, 'CITY']

    # Filter users with the same city and different user_id
    city_match = users_data[users_data['CITY'] == user_city]
    city_match = city_match[city_match.index != user_id]

    return city_match

# Example: Recommend users for the user with user_id 0 based on common cities
user_id_to_recommend = 0
city_recommendations = recommend_users_by_city(user_id_to_recommend, users_data)

# Display the recommendations based on common cities
print(f"Recommendations for user {user_id_to_recommend} based on common cities:\n")
print(city_recommendations)
```
